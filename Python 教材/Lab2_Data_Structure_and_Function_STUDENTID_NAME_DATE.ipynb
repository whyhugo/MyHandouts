### Set

Set 就像是重型機車，只能載獨一無二的女朋友

 - Set 只能儲存不重複的值

 - 使用情境：不想收集重複資料

 - set()

````python
a_set = set()
b_set = {1,2,3,2}
c_set = {1, "1"}
b_set
````

````python
type(a_set)
````

### 參數 (Parameters)

#### 指定參數

````python
def calculate_sum(number1, number2, number3):
    sum_number = 0
    sum_number += number1
    sum_number += number2
    sum_number += number3
    return sum_number

sum_number = calculate_sum(1,2,3)
sum_number
````

#### 沒有參數
 - 沒有參數的函式，通常是用於增加程式的可讀性
 - 如果不是用來增加程式的可讀性，需仔細考慮函式的必要性

````python
def calculate_sum():
    sum_number = 0
    sum_number += 1
    sum_number += 2
    sum_number += 3
    return sum_number

sum_number = calculate_sum()
sum_number
````

#### 不定參數

````python
def calculate_sum(*numbers):
    sum_number = 0
    for number in numbers:
        sum_number += number
    return sum_number

sum_number = calculate_sum(1,2,3)
sum_number
````

````python
# 一定要輸入一個數字
def calculate_sum(number_1, *numbers):
    sum_number = number_1
    for number in numbers:
        sum_number += number
    return sum_number

sum_number = calculate_sum(1,2,3)
sum_number
````

#### 預設參數
 - 遇到沒有輸入參數時，程式自動給參數
 - 預設參數須放至後方

````python
def calculate_sum(number1=1, number2=2, number3=3):
    sum_number = 0
    sum_number += number1
    sum_number += number2
    sum_number += number3
    return sum_number

sum_number = calculate_sum()
sum_number
````

### Return
 - 要傳回值時，記得要加 return，
 - 如果傳回值需要保存，記得要指定給一個變數
 - 程式遇到 return 就停止執行後面的函式內容

判斷輸入值是否為正數，是的話回傳 True，不是的話回傳 False

````python
def is_positive(x):
    if x > 0:
        return True
    return False

is_positive(-3)
````

可以同時 return 多個資料

````python
def calculate_sum(number1, number2, number3):
    sum_number = 0
    sum_number += number1
    sum_number += number2
    sum_number += number3
    return sum_number, number1

sum_number, number1 = calculate_sum(1,2,3)
print(sum_number, number1)
````

### 全域變數 (Global variable) 與 區域變數 (Local variable)
 - 函式裡面宣告的變數為區域變數，出了函式就見光死
 - 函式外的變數為全域變數，函式可以拿來用

製作一個計算平均值的函式

````python
def calculate_mean(x1, x2, x3):
    total = 0
    total += x1
    total += x2
    total += x3
    mean = total / 3
    return mean
mean = calculate_mean(1,2,3)
mean
````
